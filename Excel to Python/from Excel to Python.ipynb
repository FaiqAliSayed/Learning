```python
#After printing interim steps, my thoughts are this happens due to iteration.
#Initially when the list gets passed as [1,1,1,1] after if loop, the index points to 0 location. This gets removed with help of remove function later.
#Now when the list gets passed again the index would point to 1st location. In this case the 3rd [1] (The second 1 would be at 0th location which gets skipped and can be seen in output).

#In order to validate my thinking, I passed [2, 3, 1,1,1,1, 4, 5] and the output remains [2, 3, 1,1, 4, 5] which I was expecting. Reason being it won't be able to find [1] at index 2 and hence this won't be removed.

#Lastly, If I pass [1, 2, 3, 1,1,1,1, 4, 5] , the output will be again [2, 3, 1,1, 4, 5]. Here because it will find [1] at index 2. Hence remove same.
```
