Some handy data visualization code

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("heart.csv")
```

```python
df.head()
```

Groupby to make things interesting 

```python
chest_pain = df.groupby(by="cp").median().reset_index()
chest_pain.head()
```

```python
chest_pain.plot.bar(x="cp",y="age")
```

If you don't specify the y axis.
Remaining all variables will be shown

```python
chest_pain.plot.bar(x="cp")
```

Now we can try same code using matplot lib

```python
fig, ax = plt.subplots()
ax.bar(chest_pain["cp"],chest_pain["age"],label="age")
ax.set_xlabel("cp")
ax.legend();
```

It did same work.
However the number of lines taken were comparatively more

we can try adding colors to the above code

```python
fig, ax = plt.subplots()
ax.bar(chest_pain["cp"],chest_pain["age"],label="age",
       edgecolor = "k",color = ["red","blue","green","yellow"])
ax.set_xlabel("cp")
ax.legend();
```

Using seaborn

```python
ax = sns.barplot("cp","age",data=df)
```

```python
df.plot.scatter("age","trestbps")
```

```python
fig, ax = plt.subplots()
ax.scatter(df["age"],df["trestbps"],marker="^",
       s=60,c=df["age"],edgecolors="k",alpha=0.5)
ax.set_xlabel("age")
ax.set_ylabel("trestbps")
```

```python
sns.scatterplot("age","trestbps",hue="age",s=30,edgecolor="none",data=df)
```

Line Plots

```python
ages = df.groupby("age").median().reset_index()
ages.head()
```

```python
ages.plot.line("age","chol")
```

```python
ages.plot.line("age",["chol","trestbps"])
```

```python
fig, ax = plt.subplots()
ax.plot(ages["age"],ages["trestbps"],ls=":",lw=1.7)
ax.set_xlabel("age")
ax.set_ylabel("trestbps")
```

```python
sns.lineplot("age","trestbps",data=ages)
```
